```python
cov_data = pd.read_csv('C:/Users/OWNER/sample/3-2-3-cov.csv')
```

```python
x = cov_data['x']
y = cov_data['y']

N = len(cov_data)

mu_x = sp.mean(x)
mu_y = sp.mean(y)
```

```python
cov_sample = sum((x-mu_x)*(y-mu_y))/N
```

```python
cov_sample
```

```python
cov = sum((x-mu_x)*(y-mu_y)) / (N-1)
```

```python
cov
```

```python
sp.cov(x,y,ddof = 0)
```

```python
sp.cov(x,y,ddof = 1)
```
